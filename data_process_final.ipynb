```python
import pandas as pd

df = pd.read_csv("./Apple_tagged.csv")
df = df.drop(["Unnamed: 0"], axis = 1)
df
```

```python
# 處理內文
def comment_process(s):
    comma = ["[", "'", ']', '\\n', " " ]
    content_list = []

    for j in range(0, len(s)):
        for i in comma:
            s[j] = s[j].replace(i, "")   
        content_list.append(s[j].split(","))
#         print(len(content_list))
    return content_list
```

```python
# 刪除空白
def del_space(s):
    for i in s:
        if '' in i:
            i.remove('')
    return s
```

```python
# 依類別(func)寫入評論至list
def write_contetnt(s, d, func):
    for i in range(0, len(s)):
        for j in s[i]:
            d[func].append(j)
    return d
            
```

```python
cp_name = ['iphone5', 'iphone5s', 'iphone6', 'iphone6s', 'iphone7', 'iphone7s', 'iphone8', 'iphone8s', 'iphone11', 'iphone11Pro', 'iphone11ProMax', 'iphoneX', 'iphoneXR', 'iphoneXS', 'iphoneXsMax', 'iphoneSE', 'iphoneSE2']
```

```python
all_type = []
for name in cp_name:
    elec=[]
    cam=[]
    per=[]
    appear=[]
    a_type = dict({"Name":name, "durability":elec, 
                   "camera":cam, "performance":per, "apperance":appear})
    cam_comment = []
    elec_comment = []
    per_comment = []
    appear_comment = []
    for i in range(0, len(df["標題"])):
        if df["手機型號"][i] == "iphoneSE" or "iphoneSE" in df["手機型號2"][i]  :
            if df["續航力"][i] == 1:
                elec_comment.append(df["使用者評論"][i])
            if df["相機"][i] == 1:
                cam_comment.append(df["使用者評論"][i])
            if df["效能"][i] == 1:
                per_comment.append(df["使用者評論"][i])
            if df["外觀"][i] == 1:
                appear_comment.append(df["使用者評論"][i])   
                
    elec_comment_new = del_space(comment_process(elec_comment))
    cam_comment_new = del_space(comment_process(cam_comment))
    per_comment_new = del_space(comment_process(per_comment))
    appear_comment_new = del_space(comment_process(appear_comment))
    
    func = ['durability', 'camera', 'performance', 'apperance']
    type_ = [elec_comment_new, cam_comment_new, per_comment_new, appear_comment_new]
    for i in range(0, 4):
        a_type = write_contetnt(type_[i], a_type, func[i])

    all_type.append(a_type)
```
